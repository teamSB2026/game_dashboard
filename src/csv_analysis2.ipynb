```python
import pandas as pd
```

```python
# CSVファイルを読み込み
df = pd.read_csv('../data/班分け結果_名前付き_R1_最終版.csv', header=0)
print(df.columns)
df.head()
```

```python
import pandas as pd
import pykakasi

# pykakasiの初期化
kks = pykakasi.kakasi()

def get_reading(text):
    """日本語テキストを読み仮名（ひらがな）に変換"""
    if pd.isna(text):
        return ''
    result = kks.convert(str(text))
    return ''.join([item['hira'] for item in result])

# R1用のメンバー列（メンバー6まで）
member_columns_r1 = ['メンバー1 ', 'メンバー2', 'メンバー3', 'メンバー4', 'メンバー5', 'メンバー6']

# R1のデータを集める
all_members_r1 = []
for col in member_columns_r1:
    if col in df.columns:
        temp_df = df[df[col].notna()][['班ID (R1)', col]].copy()
        temp_df.columns = ['班ID', 'メンバー名']
        all_members_r1.append(temp_df)

# R1を結合してソート
combined_r1 = pd.concat(all_members_r1, ignore_index=True)
combined_r1['読み仮名'] = combined_r1['メンバー名'].apply(get_reading)
combined_r1 = combined_r1.sort_values(by='読み仮名').reset_index(drop=True)
r1_df = combined_r1[combined_r1['班ID'].astype(str).str.startswith('R1')][['班ID', 'メンバー名']].reset_index(drop=True)

print("R1の班:")
print(r1_df)
r1_df.to_csv('../data/班分け結果_R1_名前付き_50音順.csv', index=False, encoding='utf-8-sig')
```

```python
# CSVファイルを読み込み
df = pd.read_csv('../data/班分け結果_名前付き_R2_最終版.csv', header=0
                )
print(df.columns)
df.head()
```

```python
import pandas as pd
import pykakasi

# pykakasiの初期化
kks = pykakasi.kakasi()

def get_reading(text):
    """日本語テキストを読み仮名（ひらがな）に変換"""
    if pd.isna(text):
        return ''
    result = kks.convert(str(text))
    return ''.join([item['hira'] for item in result])

# メンバー1〜6のデータを集める
member_columns_1to6 = ['メンバー1', 'メンバー2', 'メンバー3', 'メンバー4', 'メンバー5', 'メンバー6']

all_members_1to6 = []
for col in member_columns_1to6:
    temp_df = df[df[col].notna()][['班ID (R2)', col]].copy()
    temp_df.columns = ['班ID', 'メンバー名']
    all_members_1to6.append(temp_df)
    print(f"{col}: {len(temp_df)}件のデータを取得")

# メンバー1〜6を結合
combined_r2 = pd.concat(all_members_1to6, ignore_index=True)

# メンバー7が存在する場合は追加
if 'メンバー7' in df.columns:
    temp_df_7 = df[df['メンバー7'].notna()][['班ID (R2)', 'メンバー7']].copy()
    if len(temp_df_7) > 0:
        temp_df_7.columns = ['班ID', 'メンバー名']
        combined_r2 = pd.concat([combined_r2, temp_df_7], ignore_index=True)
        print(f"メンバー7: {len(temp_df_7)}件のデータを取得")
    else:
        print("メンバー7: データなし")
else:
    print("メンバー7: 列が存在しません")

# 読み仮名を追加して50音順にソート
combined_r2['読み仮名'] = combined_r2['メンバー名'].apply(get_reading)
combined_r2 = combined_r2.sort_values(by='読み仮名').reset_index(drop=True)

# R2のみをフィルタ
r2_df = combined_r2[combined_r2['班ID'].astype(str).str.startswith('R2')][['班ID', 'メンバー名']].reset_index(drop=True)

print("\nR2の班:")
print(r2_df)
print(f"\n合計: {len(r2_df)}名")
r2_df.to_csv('../data/班分け結果_R2_名前付き_50音順.csv', index=False, encoding='utf-8-sig')
```
