```python
import pandas as pd
```

```python
# CSVファイルを読み込み
df_班分け = pd.read_csv('../data/班分け結果.csv', header=1)
df_班分け.head()
```

```python
# 別のDataFrame（名前マッピング用）を読み込む
# ここでは仮のファイルパスを使用していますが、実際のファイルパスに変更してください
df_名前マッピング = pd.read_csv('../data/名簿一覧_Bridge_ID割当済_最終版.csv')  # 実際のファイル名に変更

# Bridge IDから名前へのマッピング辞書を作成
# df_名前マッピングには「新Bridge ID」と「お名前(漢字)」の列があると仮定
bridge_id_to_name = dict(zip(df_名前マッピング['新Bridge ID'], df_名前マッピング['お名前(漢字)']))

# メンバー列を特定（メンバー1からメンバー7まで）
member_columns = ['メンバー1', 'メンバー2', 'メンバー3', 'メンバー4', 'メンバー5', 'メンバー6', 'メンバー7']

# 存在する列だけを対象にする（メンバー1 に空白があるので注意）
#existing_member_columns = [col for col in df_班分け.columns if 'メンバー' in col]

# 各メンバー列のBridge IDを名前に置き換え
for col in df_班分け.columns:
    df_班分け[col] = df_班分け[col].map(bridge_id_to_name).fillna(df_班分け[col])

# 結果を新しいCSVファイルに保存
df_班分け.to_csv('../data/班分け結果_名前付き_最終版.csv', index=False, encoding='utf-8-sig')

print("Bridge IDを名前に置き換えました。")
```
